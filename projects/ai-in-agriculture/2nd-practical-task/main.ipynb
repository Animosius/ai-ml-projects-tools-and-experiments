# 1. Making csv dataset

```python
# setup 
import random
import pandas as pd
import openpyxl as pxl
import json 
import sqlite3
```

```python
# let's make a line with noise

# 1. Set placeholder for data:
days = []
location_places = ["Section1", "Section2", "Section3"]
location = []
pH = []
temperature = []

# Run this block to reset data in runtime
```

```python
# 2 Making a noise function

def noise_introducer(x):
    return x + random.uniform(-0.25, 0.25)
```

```python
# 3. Making the dataset

for loc in location_places:
    for day in range(1, 6):
        location.append(loc)
        days.append(day)
        pH.append(noise_introducer(6.3))
        temperature.append(noise_introducer(20))

data = {
    "days": days,
    "location": location,
    "pH": pH,
    "temperature": temperature
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# 4. Exporting to csv 
df.to_csv("data.csv", index=False)
```

# 2. Reading csv dataset

# 3 Creating a json summary

```python
# Making a summary dict
# Note: We need to state dtype as well
summary = {
    "total_records": len(imported_csv_data),
    "average_pH": float(imported_csv_data["pH"].mean()),
    "average_temperature": float(imported_csv_data["temperature"].mean())
}
```

```python
# Checking list
summary
```

```python
# Exporting to json file format
with open("irrigation_summary.json", "w") as f:
    json.dump(summary, f, indent=2)
```

# 4. Storing records in a SQLite database

```python
conn = sqlite3.connect("agriculture.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS irrigation_data (
    id INTEGER PRIMARY KEY,
    day INTEGER,
    location TEXT,
    ph FLOAT,
    temperature FLOAT
               )
""")
```

```python
rows_to_insert = imported_csv_data.to_records(index=False)

cursor.executemany("""
INSERT INTO irrigation_data (day, location, ph, temperature)
VALUES (?, ?, ?, ?)
""", rows_to_insert)
conn.commit()
print("Recorsds inserted successfully.")
```
